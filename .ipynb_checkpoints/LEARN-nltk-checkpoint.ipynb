We will first go through tokenization:
For example we have a text: Above we have already downloaded the punkt
Example:

```python
text = "Hello my name is abdul rehman. I am from pakistan"
from nltk.tokenize import word_tokenize
print(word_tokenize(text))
```

		Output: 
			['Hello my name is abdul rehman.', 'I am from pakistan']
		
Here we  separated each word and put them into an array

**FreqDist library**
Example:


```python
new_text = "Hello my name is abdul rehman. I am from pakistan. I come from a family of 5. I have a cat as pet."
word_tokenized = word_tokenize(new_text)
from nltk.probability import FreqDist
fd = FreqDist(word_tokenized)
print(fd.most_common(3))
```

		Output:
			[('.', 4), ('I', 3), ('from', 2)]

Here we can use the library to find the most repeated value in the tokenized array, where when we mention print(fd.most_common(3)) we want the 3 words that most appear in the sentence.

Now for a better visualization we can use matplotlib which is a library to make graphs
For this we will install using terminal:
								Pip install matplotlib.pyplot as plt
Then the code as follows:
Example:


```python
import matplotlib.pyplot as plt
xtext = "Hello my name is abdul rehman. I am from pakistan. I come from a family of 5. I have a cat as pet."
word_tokenized = word_tokenize(xtext)
fd2 = FreqDist(word_tokenized)
print(fd2.most_common(3))

fd2.plot(30,cumulative = False) # 30 is the scale
plt.show() # To reveal the graph
```

Here we have the overall word distribution.

***Learning NLTK PART 2***


In this we will focus on stop words:
Before we start we have to download the dataset for stopwords as follows:

```python
import nltk
nltk.download("stopwords")
```

This dataset contains all the stop words in the English dictionary
Once the data is downloaded we can delete that line of code

Moving forward:
Now we can import this dataset as follows:

```python
from nltk.corpus import stopwords
```

Now we can capture these words in any kind array or set

```python
stopwords = stopwords.words('english')
print(stopwords)
```


We need to specify english so we get the stopwords only from english language


Now from a text we will remove the stop words from the text

```python
		words = word_tokenize(xtext) #note we used xtext before, therefore we are using it again
		
		without_stop_words = []
		
		for word in words:
		    if word not in stopwords:
		        without_stop_words.append(word)
		
print(without_stop_words)
```

At the end there is a very simple python snippet to remove the stopwords

We can further see the words with were removed:

```python
print(set(words) - set(without_stop_words))
```
